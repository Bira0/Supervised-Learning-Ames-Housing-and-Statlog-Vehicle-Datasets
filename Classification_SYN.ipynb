Description of the vehicle silhouettes Dataset


* Number of Instances: The dataset consists of 846 instances or observations.
* Number of Features: There are 19 different features or variables that describe various aspects of the residential properties.
* Target Variable: The target variable in the dataset is the "Car Type," representing the type or category of car associated with the property.  
* Data Types:  The dataset contains 18 numerical features and 1 categorical feature (the "Car Type" target variable).

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# #Because this a classification problem and the distance is very important
# #we will standardize all numerical features
# from sklearn.preprocessing import StandardScaler

# sc = StandardScaler()

# X_train = sc.fit_transform(X_train)
# X_test = sc.transform(X_test)
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler(feature_range=(-1, 1))
X_train = mms.fit_transform(X_train)
X_test = mms.transform(X_test)
```

```python
X_train
```

#Model Implementation

## Support Vector Machines (SVM)

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svm = SVC(C= 10, gamma=1, kernel='poly')

svm.fit(X_train, y_train)

y_pred = svm.predict(X_test)

accuracy_svm = accuracy_score(y_test, y_pred)

print("<-------------------------------------------------->")
print(f"Accuracy score   : {accuracy_svm}")
print("<-------------------------------------------------->")
print(f"Train Score: {svm.score(X_train, y_train)}")
print(f"Test Score : {svm.score(X_test, y_test)}")


a a            a
b b b         b     b
0.01
```

* Both Poly & rbf kernels handle nonlinear relationships, so we used both.

* Lower gamma (Points far apart influence each other in drawing the boundary) smoother boundary

* Higher gamma (points closer influence each other in drawing the boundary) detailed boundary

* High accuracy score, made correct predictions on almost 90% of the data.

* Perfect training score and a very good test score, indicates strong generalization.

The gap between the train and test score, still indicates overfitting.

```python
from sklearn.model_selection import GridSearchCV, KFold
gr_param={
    'C':[1, 5, 10, 20, 100],
    'kernel':['rbf', 'poly'],
    'gamma': [0.1, 0.5, 1, 2, 5, 10, 100]
    }


cv = KFold(n_splits =10, shuffle = True , random_state = 42)

gr_model = GridSearchCV(estimator = svm, param_grid = gr_param, cv = cv, scoring = 'accuracy', verbose = 1, n_jobs = -1)

gr_model.fit(X_train,y_train)
print(f"Grid parameters: { gr_model.best_params_}")
print(f"Grid accuracy score: {gr_model.best_score_} ")
```

```python
svm2=SVC(C=1, gamma=1, kernel='poly')
svm2.fit(X_train, y_train)
y_pred = svm2.predict(X_test)
accuracy_svm2 = accuracy_score(y_test, y_pred)

print("<-------------------------------------------------->")
print(f"Accuracy score   : {accuracy_svm2}")
print("<-------------------------------------------------->")
print(f"Train Score: {svm2.score(X_train, y_train)}")
print(f"Test Score : {svm2.score(X_test, y_test)}")
```

All scores went down a little, but the result is a more balanced generalization score because the gap between the test & train scores is smaller.

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lgstc = LogisticRegression(C = 0.1,  max_iter=2000, solver='liblinear', penalty='l1')
lgstc.fit(X_train, y_train)

y_pred = lgstc.predict(X_test)
accuracy_lgstc = accuracy_score(y_test, y_pred)

print("<-------------------------------------------------->")
print(f"Accuracy score   : {accuracy_lgstc}")
print("<-------------------------------------------------->")
print(f"Train Score: {lgstc.score(X_train, y_train)}")
print(f"Test Score : {lgstc.score(X_test, y_test)}")
```

* Very bad accuracy score

* Both train & test scores are really low, indication of underfitting because of how close and bad they are.

```python
from sklearn.model_selection import GridSearchCV , KFold

param_logistic = {
    'C' : [0.1, 0.01, 0.001, 1, 10, 50, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear'],
    'max_iter': [2000, 10000, 20000, 50000, 100000]

}


#notice random_state to keep same shuffle as train_test_split and n_splits the number of folds
cv = KFold(n_splits= 5, shuffle= True, random_state= 42)

gridSearch = GridSearchCV(estimator = lgstc, param_grid= param_logistic, cv=cv,
scoring= 'accuracy', verbose= 2, n_jobs= -1)

gridSearch.fit(X_train, y_train)

print(f"Grid Parameters: {gridSearch.best_params_}")
print(f"Grid accuracy Score: {gridSearch.best_score_}")
```

```python
logreg = LogisticRegression(C = 10,  max_iter=2000, solver='liblinear', penalty='l1')
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
accuracy_logreg = accuracy_score(y_test, y_pred)

print("<-------------------------------------------------->")
print(f"Accuracy score   : {accuracy_logreg}")
print("<-------------------------------------------------->")
print(f"Train Score: {logreg.score(X_train, y_train)}")
print(f"Test Score : {logreg.score(X_test, y_test)}")
```

* Correctly classified 81% of our data

* Both the train & test scores are almost identical and they both retain a high score, suggesting that the model generalizes well without any overfitting.

* The model performs similarly on both the training and test sets, with a small difference between the train and test scores. This indicates that the model is generalizing well and does not suffer from overfitting.

## Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dtc = DecisionTreeClassifier(criterion='gini',max_depth=50, max_features=None, max_leaf_nodes=None, min_samples_leaf=10, min_samples_split=5)

dtc.fit(X_train, y_train)

y_pred = dtc.predict(X_test)
accuracy_dtc = accuracy_score(y_test, y_pred)

print("<-------------------------------------------------->")
print(f"Accuracy score   : {accuracy_dtc}")
print("<-------------------------------------------------->")
print(f"Train Score: {dtc.score(X_train, y_train)}")
print(f"Test Score : {dtc.score(X_test, y_test)}")
```

* Gini checks how often a wrong guess would be made at a node
* Entropy checks how mixed or disordered a node is
** Gini is quicker, entropy is more detailed

* Low test score, bad generalization.
* Higher train score, indication of clear overfitting.

```python
from sklearn.model_selection import GridSearchCV, KFold

param_dtc = {
             'max_depth': [ 5, 10, 20, 30, 50],
             'max_features': [None, 'sqrt', 'log2'],
             'max_leaf_nodes': [5, 10, 20, 30],
             'min_samples_split': [2, 5, 10, 20],
             'min_samples_leaf': [1, 2, 3, 4, 5, 10],
             'criterion': ['gini', 'entropy']
             }

cv = KFold(n_splits = 5, shuffle = True , random_state = 42)

gridSearch = GridSearchCV(estimator = dtc, param_grid = param_dtc, cv = cv, scoring = 'accuracy', verbose = 1, n_jobs = -1)
gridSearch.fit(X_train, y_train)

print(f"Grid Parameters: {gridSearch.best_params_}")
print(f"Grid Accuracy Score: {gridSearch.best_score_}")
```

```python
dtc = DecisionTreeClassifier(criterion='gini', max_depth=50, max_features='log2', max_leaf_nodes=20, min_samples_leaf=3, min_samples_split=10)

dtc.fit(X_train, y_train)

y_pred = dtc.predict(X_test)
accuracy_dtc2 = accuracy_score(y_test, y_pred)

print("<-------------------------------------------------->")
print(f"Accuracy score   : {accuracy_dtc2}")
print("<-------------------------------------------------->")
print(f"Train Score: {dtc.score(X_train, y_train)}")
print(f"Test Score : {dtc.score(X_test, y_test)}")
```

####Compare model performance before and after

```python
print(f"Accuracy DecisionTree Model before: {accuracy_dtc}")
print("<-------------------------------------------------->")
print(f"Accuracy DecisionTree Model after: {accuracy_dtc2}")
```

* The decrease in accuracy from 70% to 66% indicates that the adjustments made to the Decision Tree model resulted in a performance drop. This suggests that the changes may have caused the model to lose its ability to generalize well to the test data. Further tuning or evaluation might be needed to improve the model's performance.

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rfc = RandomForestClassifier(max_features=None, criterion='gini',min_samples_leaf=5,min_samples_split=10,n_estimators=10, max_depth=20, random_state=42, bootstrap=True)
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)
accuracy_rfc = accuracy_score(y_test, y_pred)

print("<-------------------------------------------------->")
print(f"Accuracy score   : {accuracy_rfc}")
print("<-------------------------------------------------->")
print(f"Train Score: {rfc.score(X_train, y_train)}")
print(f"Test Score : {rfc.score(X_test, y_test)}")
```

* Accuracy score, a bad score of [0.76] indicates that the model doesn't make that many predictions.

* Train score [0.91] is very high compared to the test score [0.76] suggesting a large amount of overfitting.

```python
from sklearn.model_selection import GridSearchCV, KFold

rfc_grid = RandomForestClassifier(random_state=42)

# Define the grid of hyperparameters
param_frst = {
    'n_estimators': [20, 50, 100],
    'max_depth': [None, 10, 20, 30, 40],
    'bootstrap': [True, False],
    'max_features': [None,'sqrt', 'log2'],
}

cv = KFold(n_splits = 10, shuffle = True , random_state = 42)

grid_search = GridSearchCV(
    estimator=rfc_grid,
     param_grid=param_frst,
     cv=cv,
     scoring='accuracy',
     verbose=3,
     n_jobs=-1
 )

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

Bootstrap splits the training dataset into smaller datasets, and trains on them one by one.

```python
rfc2 = RandomForestClassifier(bootstrap=False, max_depth=None, max_features='sqrt', n_estimators=100)

rfc2.fit(X_train, y_train)

y_pred = rfc2.predict(X_test)
accuracy_rfc2 = accuracy_score(y_test, y_pred)

print("<-------------------------------------------------->")
print(f"Accuracy score   : {accuracy_rfc2}")
print("<-------------------------------------------------->")
print(f"Train Score: {rfc2.score(X_train, y_train)}")
print(f"Test Score : {rfc2.score(X_test, y_test)}")

```

## K-Nearest Neighbors (K-NN)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

KNN_class = KNeighborsClassifier(n_neighbors=6, metric='euclidean')
KNN_class.fit(X_train, y_train)

y_pred = KNN_class.predict(X_test)
accuracy_KNN = accuracy_score(y_test, y_pred)


print("<-------------------------------------------------->")
print(f"Accuracy score   : {accuracy_KNN}")
print("<-------------------------------------------------->")
print(f"Train Score: {KNN_class.score(X_train, y_train)}")
print(f"Test Score : {KNN_class.score(X_test, y_test)}")
```

* Low Accuracy score, indicates that the model doesn't make many predictions.
* Close train test scores, but the scores are not high enough to our liking, needs more tuning.

```python
from sklearn.model_selection import GridSearchCV, KFold

param_KNN = {
    'n_neighbors': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'metric': ['euclidean', 'manhattan']
}

#notice random_state to keep same shuffle as train_test_split
cv = KFold(n_splits = 5, shuffle = True , random_state = 42)

gridSearch = GridSearchCV(estimator = KNN_class, param_grid = param_KNN, cv = cv, scoring = 'accuracy', verbose = 2, n_jobs = -1)
gridSearch.fit(X_train, y_train)

print(f"Grid Parameters: {gridSearch.best_params_}")
print(f"Grid Accuracy Score: {gridSearch.best_score_}")
```

Using suggested 'Manhattan' distance metric and 6 Neighbours

```python
KNN_class = KNeighborsClassifier(n_neighbors = 6, metric='manhattan')
KNN_class.fit(X_train, y_train)

y_pred = KNN_class.predict(X_test)
accuracy_KNN2 = accuracy_score(y_test, y_pred)


print("<-------------------------------------------------->")
print(f"Accuracy score   : {accuracy_KNN2}")
print("<-------------------------------------------------->")
print(f"Train Score: {KNN_class.score(X_train, y_train)}")
print(f"Test Score : {KNN_class.score(X_test, y_test)}")
```
